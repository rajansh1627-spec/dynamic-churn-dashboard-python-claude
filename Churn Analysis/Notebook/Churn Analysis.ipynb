```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Churn.csv")
data.head(5)
```

```python
data.info()
```

```python
data['Purchase_Date']=pd.to_datetime(data['Purchase_Date'],errors='coerce')
data['Customer_ID']=data['Customer_ID'].astype('Int64')
data['Age']=data['Age'].astype('Int64')
```

```python
data['Churn']=data['Churn'].replace({'Yes':1 ,'No':0})
```

```python
data['Age'] = data['Age'].replace([-1, 200], np.nan)
data['Age']=data['Age'].fillna(data['Age'].median(), inplace=True)
data['Satisfaction_Score']=data['Satisfaction_Score'].fillna(data['Satisfaction_Score'].median(), inplace=True)
data['Churn']=data['Churn'].fillna(data['Churn'].mode()[0], inplace=True)
```

```python
data['Gender']=data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)
data['Country']=data['Country'].fillna(data['Country'].mode()[0], inplace=True)
data['Product_Category']=data['Product_Category'].fillna(data['Product_Category'].mode()[0], inplace=True)
data['Purchase_Date'] = data['Purchase_Date'].ffill()
```

```python
data['Purchase_Amount']=data['Purchase_Amount'].abs()
```

```python
data['Country']=data['Country'].replace({'U.S.A':'USA','Ind':'India'})

data['Gender']=data['Gender'].replace({'F':'Female','M':'Male'})

data['Product_Category']=data['Product_Category'].replace({'Elec':'Electronics'})

data['Satisfaction_Score']=data['Satisfaction_Score'].replace({10:5})
```

```python
data.dropna(subset=['Customer_ID'], inplace=True)
```
